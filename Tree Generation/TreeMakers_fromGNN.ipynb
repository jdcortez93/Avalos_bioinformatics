```python
import pandas as pd
from Bio import SeqIO #Reading and Writing Biological Sequences
import subprocess #to run linux with pythonic style
from io import StringIO
import sys
import os, fnmatch
from datetime import datetime

def find(pattern, path):
    result = []
    for root, dirs, files in os.walk(path):
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                result.append(os.path.join(root, name))
    return result



# Path for using and dumping data
home_path = '/Users/jeremydcortez/Bioconda_Playground/25-0219/003/selectedGNN0/'
os.chdir(home_path)

# Print current path
path = os.getcwd()
print("Start path is now "+path)
```

```python
# File to operate on, must be name before .fasta
new = 'ID_selected0220' # Change file name! Do NOT type .fasta
os.makedirs('output', exist_ok=True)
```

```python
# Execute a command to mafft
mafftoutput = new+'_mafft.fasta'
os.chdir(home_path+'output')
with open('outputlog'+str(new)+'.txt', "w") as f:
    print('')
    print('------------------- Running MAFFT -------------------')
    print('')
    subprocess.call(['mafft --auto --reorder '+home_path+new+'.fasta > '+mafftoutput], shell=True, stdout=f, stderr=f)
```

```python
# Execute a command to BMGE
with open('outputlog'+str(new)+'.txt', "a") as f:
    print('')
    print(' ------------------ Running BMGE ------------------')
    print('')
    subprocess.run(['bmge -i '+new+'_mafft.fasta -t AA -o '+new+'_mafft_bmge.fasta'], shell=True, stdout=f, stderr=f)
```

```python
# Execute a command to FastTree
with open('outputlog'+str(new)+'.txt', "a") as f:
    print('')
    print('-----------  RUNNING FASTREE  -----------')
    print('')
    subprocess.run(['FastTree '+new+'_mafft_bmge.fasta > '+new+'_mafft_bmge_fasttree.txt'], shell=True, stdout=f, stderr=f)

    # output is in Newick format
```
